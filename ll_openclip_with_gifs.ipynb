````python
def promt_based_rwd(clip_sim):
    rwd = {0: 100, 1: -1, 2: -100}
    return rwd[np.argmax(clip_sim)]
````

````python
class CLIPEnv():
    def __init__(self, env, clip_model, clip_preprocess, tokenizer, prompt, writer):
        self.env = env
        self.model = clip_model
        self.preprocess = clip_preprocess
        self.tokenizer = tokenizer
        text = self.tokenizer(prompt)
        if  torch.cuda.is_available():
          text = text.cuda()
        self.text_features = self.model.encode_text(text)
        self.text_features /= self.text_features.norm(dim=-1, keepdim=True)
       
        self.action_space = self.env.action_space
        self.observation_space = self.env.observation_space
        self.metadata = self.env.metadata
        self.clip_rewards_per_episode = []
        self.env_rewards_per_episode = []
        self.disc_clip_rewards_per_episode = []

        
        self.clip_rewards = []
        self.env_rewards = []
        self.ep_lenghts = [0]
        self.disc_clip_rewards = []
        
        self.writer = writer
        self.n_steps = 0
        self.n_episodes = 0

        self.frames = []

    def render(self):
        return self.env.render(mode="rgb_array")
    
    def reset(self):
        return self.env.reset()
    
    def close(self):
        return self.env.close()
    
    def step(self, action):
        next_st, rwd, done, info = self.env.step(action)
        img = self.env.render(mode="rgb_array")
        clip_rwd = self.get_clip_reward(img)
        
        self.clip_rewards_per_episode.append(clip_rwd)
        self.env_rewards_per_episode.append(rwd)

        mean_clip_rwd = np.mean(
            self.clip_rewards_per_episode[-min(9, len(self.clip_rewards_per_episode)):]  + [clip_rwd])
        
        disc_clip_rwd = promt_based_rwd(clip_rwd)
        
        
        self.disc_clip_rewards_per_episode.append(disc_clip_rwd)

        if done:
            if self.writer:
                self.writer.add_scalar('episode_rewards/env_reward',  sum(self.env_rewards_per_episode), self.n_episodes)
                # self.writer.add_scalar('episode_rewards/clip_reward', sum(self.clip_rewards_per_episode) , self.n_episodes)
                # self.writer.add_scalar('episode_rewards/mean_per_episode_clip_reward', np.mean(self.clip_rewards_per_episode) , self.n_episodes)

                self.writer.add_scalar('episode_rewards/discrete_clip_reward', np.sum(self.disc_clip_rewards_per_episode) , self.n_episodes)

                self.writer.add_scalar('Episode length', self.n_steps  - self.ep_lenghts[-1], self.n_episodes)

            self.env_rewards.append(self.env_rewards_per_episode)
            self.clip_rewards.append(self.clip_rewards_per_episode)
            self.disc_clip_rewards.append(self.disc_clip_rewards_per_episode)


            self.env_rewards_per_episode = []
            self.clip_rewards_per_episode = []
            self.disc_clip_rewards_per_episode = []

            
            self.n_episodes += 1
            self.ep_lenghts.append(self.n_steps)
            self.frames = []
        
        self.n_steps += 1
        
        return next_st, disc_clip_rwd, done, info
    
    def get_clip_reward(self, state):
        image = self.preprocess(Image.fromarray(np.uint8(state))).unsqueeze(0)
        if torch.cuda.is_available():
          image = image.cuda()
        with torch.no_grad(), torch.cuda.amp.autocast():
            image_features = self.model.encode_image(image)
            image_features /= image_features.norm(dim=-1, keepdim=True)
            self.frames.append(image_features)
            if len(self.frames) > 5:
                self.frames.pop(0)
            mean_image_features = torch.unsqueeze(torch.mean(torch.concatenate(self.frames), dim=0), dim=0)
            sim = (mean_image_features @ self.text_features.T)
        return sim.cpu().detach().numpy()[0]
    @property
    def unwrapped(self):
        return self.env.unwrapped
````

````python
def run_exp(agent, env, prompt, clip_model_name, env_name, exp_path, n_steps, notes=''):
    if not os.path.exists(exp_path):
        os.mkdir(exp_path)

    
    agent.learn(total_timesteps=n_steps, progress_bar=True)
    agent.save(f"{exp_path}/agent")
    
    corr = stats.pearsonr([sum(e) for e in env.env_rewards], [np.mean(e) for e in env.disc_clip_rewards])[0]
    m_rwd = np.mean([sum(e) for e in env.env_rewards[-10:]])
    results = {
        'env_name': env_name,
        'prompt': prompt,
        'clip_model_name': clip_model_name,
        'correlation': corr,
        'mean_env_rwd_over_last_10_episodes': m_rwd,
        'n_episodes': env.n_episodes,
        'n_steps': env.n_steps,
         'notes': notes,
    }
    with open(f'{exp_path}/results.json', 'w') as f:
        json.dump(results, f)
    
    # compute correlation between env and clip rewards for each episode separately
    per_episode_corr = [stats.pearsonr(e, c)[0] for e, c in zip(env.env_rewards, env.disc_clip_rewards)]
    for i in range(env.n_episodes):
        env.writer.add_scalar('Per episode correlation', per_episode_corr[i], i)
````

````python
model, _, preprocess = open_clip.create_model_and_transforms('ViT-B-32-quickgelu', pretrained='laion400m_e32')
tokenizer = open_clip.get_tokenizer('ViT-B-32-quickgelu')
if torch.cuda.is_available():
    model = model.cuda()
````

````python
EXP_PATH = 'experiments/colab_3_prompts_clip_rwd_mean_frame_400k_steps/'
ENV_NAME = 'LunarLander-v2'
N_STEPS = 400000

env = gym.make(ENV_NAME)
prompt =  ['a video game of a rocket on the snow between yellow flags', 'a video game of a flying rocket far from the snow', 
    'a ski slope with two flags above it and a rocket on the ski slope to the right of the flags'
    ]
writer = SummaryWriter(EXP_PATH)

cl_env = CLIPEnv(env, model, preprocess, tokenizer, prompt, writer)


agent = DQN('MlpPolicy', cl_env, verbose=0, learning_rate=1e-3,learning_starts=1000, buffer_size=50000, target_update_interval=1000)
````

````python
run_exp(agent, cl_env, prompt, 'open_clip_ViT-B-32-quickgelu_laion400m_e32', ENV_NAME, EXP_PATH, N_STEPS,
       'corr btw sum of env rwds and sum of dics clip rwds, train on clip rwd from 3 prompts with mean frame embds over 5, learning_rate=1e-3,learning_starts=1000, buffer_size=50000, target_update_interval=1000')
````

````python
from skvideo import io

def record_agent(env, agent, n_eps, exp_path, save_images=False):
    for i in range(n_eps):
        obs = env.reset()
        done = False
        frames = [ ]
        if not os.path.exists(exp_path + '/frames'):
            os.mkdir(exp_path + '/frames')
        while not done:
            action, _states = agent.predict(obs)
            obs, rewards, done, info = env.step(action)
            img = env.render()
            if save_images:
                im = Image.fromarray(img)
                im.save(exp_path + f'/frames/frame_{len(frames)}.png')
            frames.append(img)
        io.vwrite(exp_path + f'/play_episode_{i}.mp4', np.array(frames))
````

````python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
````
